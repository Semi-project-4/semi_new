#### ìŠ¤ì¼€ì¼ë§ë³„ íŠ¹ì§•

|-|-|
|í‘œì¤€í™” ìŠ¤ì¼€ì¼ë§|ì•„ì›ƒë¼ì´ì–´ì˜ ì˜í–¥ì„ ë°›ìŒ.<br>í‰ê·  0, ë¶„ì‚° 1ì´ ë˜ê²Œë” ë¶„í¬ì‹œí‚¤ê¸° ë•Œë¬¸ì—, **ë°ì´í„°ì˜ ê¸°ì¡´ ë¶„í¬ í˜•íƒœê°€ ì‚¬ë¼ì§€ê³ ** ì •ê·œë¶„í¬ë¥¼ ë”°ë¥´ëŠ” ê²°ê³¼ë¬¼ì„ ê°€ì ¸ ì˜´|

|ìµœì†Œ-ìµœëŒ€ ìŠ¤ì¼€ì¼ë§|ì•„ì›ƒë¼ì´ì–´ì˜ ì˜í–¥ì„ ë°›ìŒ.<br>ìœ„ì˜ ë‘ ìŠ¤ì¼€ì¼ëŸ¬ì™€ ë¹„êµí–ˆì„ ë•Œ, **ë°ì´í„°ì˜ ê¸°ì¡´ ë¶„í¬ë¥¼ ê°€ìž¥ ìžˆëŠ” ê·¸ëŒ€ë¡œ ë‹´ì•„ë‚´ë©° ìŠ¤ì¼€ì¼ë§Œ ë³€í™”**ì‹œí‚´.<br>ë°ì´í„°ì˜ ë²”ìœ„ëŠ” 0~1ë¡œ ë‚˜íƒ€ë‚¨|

#### ìŠ¤ì¼€ì¼ë§ ì ìš© ì‹œ ì£¼ì˜ì 
* ìŠ¤ì¼€ì¼ë§ ëŒ€ìƒì—ì„œ ì¢…ì†ë³€ìˆ˜ë¥¼ ì œì™¸
* ìŠ¤ì¼€ì¼ë§ ì „ í›ˆë ¨ì…‹ê³¼ ì‹œí—˜ì…‹ì„ êµ¬ë¶„

````python
# ì¢…ì†ë³€ìˆ˜ (y) / ë…ë¦½ë³€ìˆ˜ë“¤ (x)
# í›ˆë ¨ì…‹ê³¼ ì‹œí—˜ì…‹ êµ¬ë¶„ (train, test) => ìŠ¤ì¼€ì¼ë§
# X_test ìŠ¤ì¼€ì¼ë§ì€ ì–´ë–»ê²Œ í•´ìš”?
from sklearn.model_selection import train_test_split # í›ˆë ¨ì…‹/ì‹œí—˜ì…‹

# Class
X_train, X_test, y_train, y_test = train_test_split(
    df.drop('Class', axis=1), # X (ë…ë¦½ë³€ìˆ˜)
    df.Class, test_size=0.2, random_state=19
)
````

````python
X_train.head() # ìŠ¤ì¼€ì¼ë§ ëŒ€ìƒ
````

````python
mm_scaler = MinMaxScaler()
X_train_scaled = mm_scaler.fit_transform(X_train)
X_train_scaled
````

````python
# mm_scaler.min_
mm_scaler.data_max_
mm_scaler.data_min_
````

````python
# scalerì˜ fit ìžì²´ëŠ” train ë°ì´í„°ë¡œë§Œ ì§„í–‰ì„ í•˜ëŠ”ë°
# modelì„ ì‚¬ìš©í•´ì„œ pred. => test ë°ì´í„°ëŠ” ì´ë¯¸ í•™ìŠµëœ scalerë¥¼ ì‚¬ìš©í•œë‹¤
X_test_scaled = mm_scaler.transform(X_test)
X_test_scaled
````

# ëª¨ë¸ë§ & ì˜ˆì¸¡/í‰ê°€

````python
# from sklearn.linear_model...
# ì´ë¯¸ ì •ì˜ëœ ëª¨ë¸ ì•Œê³ ë¦¬ì¦˜ì„ SKlearn í˜¹ì€ ë‹¤ë¥¸ ì•Œê³ ë¦¬ì¦˜ íŒ¨í‚¤ì§€
from sklearn.neighbors import KNeighborsClassifier # ë¶„ë¥˜ ë¬¸ì œ í’€ ë•Œ (ë²”ì£¼í˜• ë³€ìˆ˜)
# from sklearn.neighbors import KNeighborsRegressor -> íšŒê·€ ë¬¸ì œ í’€ ë•Œ (ì—°ì†í˜• ë³€ìˆ˜ í’€ ë•Œ)
````

````python
knn = KNeighborsClassifier()
knn.fit(X_train_scaled, y_train)
pred = knn.predict(X_test_scaled)
````

````python
knn.predict_proba(X_test_scaled) # 5ê°œë¥¼ ë¹„êµ
````

````python
pred
````

````python
from sklearn.metrics import accuracy_score

accuracy_score(y_test, pred)
````

# í•˜ì´í¼íŒŒë¼ë¯¸í„° íŠœë‹
`KNeighborsClassifier` : ë¶„ë¥˜ ë¬¸ì œ í•´ê²° ì‹œ ì‚¬ìš©í•˜ëŠ” í´ëž˜ìŠ¤<br>
`KNeighborsRegressor` : íšŒê·€ ë¬¸ì œ í•´ê²° ì‹œ ì‚¬ìš©í•˜ëŠ” í´ëž˜ìŠ¤
```
(class) KNeighborsClassifier(
    n_neighbors: int = 5, *,
    weights: str = "uniform",
    algorithm: str = "auto",
    leaf_size: int = 30,
    p: int = 2,
    metric: str = "minkowski",
    metric_params: Any | None = None,
    n_jobs: Any | None = None)
```
* n_neighbors : ì˜ˆì¸¡ì— ì°¸ê³ í•  ì´ì›ƒ ìˆ˜ (5ê°œ)
* weights : ì˜ˆì¸¡ì— ì‚¬ìš©ë˜ëŠ” ê°€ì¤‘ì¹˜ í•¨ìˆ˜ - ê¸°ë³¸ê°’ : uniform - ëª¨ë“  ì ì— ë™ì¼í•œ ê°€ì¤‘ì¹˜ ë¶€ì—¬
* metric : ê±°ë¦¬ ì¸¡ì • ê¸°ì¤€ (minkowski)
* n_jobs : ì‹¤í–‰í•  ë³‘ë ¬ ìž‘ì—… ìˆ˜

````python
def tuning(n: int):
    '''
    nì„ ë°›ì•„ì„œ í•´ë‹¹ ê°’ìœ¼ë¡œ KNN ëª¨ë¸ì„ í•˜ì´í¼ íŒ¨ëŸ¬ë¯¸í„° ë³€ê²½
    '''
    knn = KNeighborsClassifier(n_neighbors=n) # nì˜ ê°’ì„ ë„£ìœ¼ë©´ í•´ë‹¹ nì„ knnì˜ ê·¼ì ‘ ì´ì›ƒìˆ˜ íŒë‹¨ ê¸°ì¤€ìœ¼ë¡œ ì‚¬ìš©í•˜ëŠ” ëª¨ë¸ì´ ë§Œë“¤ì–´ì§
    knn.fit(X_train_scaled, y_train)
    pred = knn.predict(X_test_scaled)
    score = accuracy_score(y_test, pred)
    return score
````

````python
tuning(7)
````

````python
for i in range(1, 21):
    print(i, tuning(i))
````

````python
scores = [(i, tuning(i)) for i in range(1, 21)]
scores
````

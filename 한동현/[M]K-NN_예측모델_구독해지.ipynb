> ##### D.모델링/1.데이터 초기회 & 전처리

  ```
- 불필요 컬럼 삭제:
  - 'user_id'
- 수치형 데이터:  StandardScaler 스케일링
- 범주형 인코딩: Low == 0, Medium == 1, High == 2, Basic == 0, Premium == 1
  ```

````python
'''3.모델링/1.데이터 초기화 & 전처리'''

# 프로세스 진행 결과 확인 함수 선언
def result_check():
  enter()

  print('[train_data]')

  enter()

  display(train_data)

  enter()

  print('[test_data]')

  enter()

  display(test_data)

  enter()


# 데이터 초기화
train_data = origin_train_data.copy()

test_data = origin_test_data.copy()

sample_submission = origin_sample_submission.copy()

print('데이터 초기화')
result_check()



# 학습, 테스트 데이터 초기화 확인 출력
print('학습, 테스트 데이터 초기화')
result_check()


# 범주형 변수 구분
categorical_cols = []

for col in train_data.columns:

    if train_data[col].dtype == 'object' and col not in ['user_id']:

        categorical_cols.append(col)

# 수치형 변수 구분
numerical_cols = []

for col in train_data.columns:

    if train_data[col].dtype in ['int64', 'float64'] and col not in ['target']:

        numerical_cols.append(col)


enter()
print("범주형 변수('user_id'제외):", (categorical_cols))
enter()
print('수치형 변수:', (numerical_cols))
enter()


# 데이터 스케일링을 위한 StandardScaler 인스턴스 생성
scaler = StandardScaler()





# 수치형 데이터에 대해 스케일링 적용
train_data[numerical_cols] = scaler.fit_transform(train_data[numerical_cols])

test_data[numerical_cols] = scaler.transform(test_data[numerical_cols])

print('데이터 스케일링')
result_check()


# 불필요한 컬럼 삭제
train_data.drop(['user_id'], axis=1, inplace=True)

test_data.drop(['user_id'], axis=1, inplace=True)


# 불필요 컬럼 삭제 출력
print('불필요 컬럼 삭제')
result_check()



# 수치형 데이터 스케일링 적용 출력
print('수치형 데이터 스케일링 적용')
result_check()



# 범주형 인코딩-학습 레벨
train_data['preferred_difficulty_level'] = train_data['preferred_difficulty_level'].map({'Low':0, 'Medium':1,'High':2})

test_data['preferred_difficulty_level'] = test_data['preferred_difficulty_level'].map({'Low':0, 'Medium':1,'High':2})


# 범주형 인코딩-구독 유형
train_data['subscription_type'] = train_data['subscription_type'].map({'Basic':0, 'Premium':1})

test_data['subscription_type'] = test_data['subscription_type'].map({'Basic':0, 'Premium':1})

print('인코딩 적용')
result_check()
````

> ##### D.모델링/2.KNN model 적용

```
- 데이터 분할
  - test_size = 0.2, rs = 12
- K-NN 모델링
  - K = 5
```

````python
'''3.모델링/2.KNN model 적용'''

# X, y 할당하기
X = train_data.drop(['target'], axis=1) # X = 구독 해지 열만 제외

y = train_data['target'] # y = 구독 해지 열만 선택


# 할당 데이터 확인
print(X.shape)

display(X)

print(y.shape)

display(y)

print(test_data.shape)

y.value_counts()


# 데이터 분할
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state=12)


# KNN 모델 생성
knn = KNeighborsClassifier(n_neighbors=5)


# 모델 학습
knn.fit(X_train, y_train)


# 테스트 데이터로 예측
y_pred = knn.predict(X_test)
````

```
- 정확도 출력: 0.557
- 오차 행렬 출력:
```

````python
'''D.모델링/3.모델 평가'''


# 정확도 평가
accuracy = accuracy_score(y_test, y_pred)
print("정확도:", accuracy)

# 클래스별 성능 지표 출력
print(classification_report(y_test, y_pred))

# 오차행렬 출력
confusion_mat = confusion_matrix(y_test, y_pred)
print("오차행렬:")
print(confusion_mat)


# 오차행렬 시각화
plt.figure(figsize=(10, 7))
sns.heatmap(confusion_mat, annot=True, fmt='g')
plt.xlabel('Predicted')
plt.ylabel('True')
plt.show()

````
